```python
env = holodeck.make('UrbanCity-MaxDistance')
# The UAV takes 3 torques and a thrust as a command.
command = np.array([0, 0, 50, 100])

env.reset()
for _ in range(500):
   state, reward, terminal, info = env.step(command)
   print(state['LocationSensor'])
```
